## EN4 Met Office Dataset

EN4 dataset is made up of monthly sub-surface profiles of temperature and salinity in the global ocean (1900 - Present). The data can be accessed from the following link, 

https://www.metoffice.gov.uk/hadobs/en4/download-en4-2-2.html

```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ppdir = "../../../Data/EN.4.2.2.analyses.g10.2009/"

ds = xr.open_mfdataset(ppdir + "*nc")

print(ds)
```

```python
fig = plt.figure(figsize=[18,4])

plt.subplot(1,2,1)
ds['temperature'].isel(time=0, depth=0).plot(robust=True)

plt.subplot(1,2,2)
ds['salinity'].isel(time=0, depth=0).plot(robust=True)
```

```python
ds['salinity'].isel(time=4, depth=1).sel(lat=slice(40, 70), lon = slice(0,50)).plot()
```

## NCAR CORE

NCAR CORE2 dataset provides surface air-sea fluxes of heat and freshwater, and ocean surface wind stress fields between 1949 and 2006. The dataset can be using the following link,

https://climatedataguide.ucar.edu/climate-data/corev2-air-sea-surface-fluxes

```python
ppdir = "../../../Data/NCAR_CORE2_Surface_Flux/"

ds = xr.open_dataset(ppdir + "2006.nc", decode_times=False)

print(ds)
```

```python
fig = plt.figure(figsize=[18,4])

plt.subplot(1,3,1)
ds['F_prec'].mean('time').plot(robust=True, cmap='Reds')

plt.subplot(1,3,2)
ds['F_evap'].mean('time').plot(robust=True, cmap='Blues')

plt.subplot(1,3,3)
ds['F_roff'].mean('time').plot(robust=True, cmap='Reds')
```
